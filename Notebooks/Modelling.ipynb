```python
import pandas as pd
import numpy as np
import pickle
```

```python
data= pd.read_csv("C://Users/hugoa/Ironhack/Week_5/Mid_bootcamp_project/Data/Cleaned/clean_extra_rows.csv", index_col=0)
```

```python
def split(x):
    
    categorical = x.select_dtypes("object")
    numerical= x.select_dtypes("int64", "float64")
    
    categorical = pd.DataFrame(categorical)
    numerical = pd.DataFrame(numerical)
    return categorical, numerical
```

```python
categorical, numerical = split(data)
```

```python
categorical.head()
```

```python
categorical= categorical[[ 'Certificate','Genre', 'lang', 'reg']]
```

```python
dummies= pd.get_dummies(categorical, drop_first=True)
dummies
with open("../encoders/dummies.pkl", "wb") as file:
    pickle.dump(dummies, file)
```

```python
numerical= numerical[[ 'Runtime','No_of_Votes', 'Gross']]
```

```python
# Defining X and Y
```

```python
X_num = numerical
y = data['IMDB_Rating']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
X_train_num, X_test_num, y_train, y_test = train_test_split(X_num, y, test_size=0.3, random_state=31)
```

```python
X_train_cat, X_test_cat= train_test_split(dummies, test_size=0.3, random_state=31)
```

```python
X_test_num=X_test_num.sort_index()
X_train_num=X_train_num.sort_index()
```

```python
transformer = MinMaxScaler().fit(X_test_new) 
with open("../transformers/transformer_test.pkl", "wb") as file:
    pickle.dump(transformer, file)
X_test_normalized = transformer.transform(X_test_new)
X_test_new_scaled_df = pd.DataFrame(X_test_normalized, columns=X_test_new.columns)
X_test_new_scaled_df
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
import math
lm = linear_model.LinearRegression()
with open("../models/lm_low_r2.pkl", "wb") as file:
    pickle.dump(lm, file)
a=lm.fit(X_train_new_scaled_df, y_train)
y_pred_train= lm.predict(X_train_new_scaled_df)
y_pred_test= lm.predict(X_test_new_scaled_df)
```

```python
print("Mean squared error (train): %.2f" % mean_squared_error(y_train, y_pred_train))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination (train): %.2f" % r2_score(y_train, y_pred_train))

print("Mean absolute error (train): %.2f" % mean_absolute_error(y_train, y_pred_train))


print("Mean squared error (test): %.2f" % mean_squared_error(y_test, y_pred_test))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination (test): %.2f" % r2_score(y_test, y_pred_test))

print("Mean absolute error (test): %.2f" % mean_absolute_error(y_test, y_pred_test))
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt

neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
  
# Loop over K values
for i, k in enumerate(neighbors):
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train_new_scaled_df, y_train)
      
    # Compute training and test data accuracy
    train_accuracy[i] = knn.score(X_train_new_scaled_df, y_train)
    test_accuracy[i] = knn.score(X_test_new_scaled_df, y_test)

plt.plot(neighbors, test_accuracy, label = 'Testing dataset Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training dataset Accuracy')
  
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
n_neighbors = 6
 
knn = KNeighborsRegressor(n_neighbors)
knn.fit(X_train_new_scaled_df, y_train)
print('Score of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train_new_scaled_df, y_train)))
print('Score of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test_new_scaled_df, y_test)))
```
