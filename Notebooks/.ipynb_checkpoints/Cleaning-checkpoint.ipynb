```python
# Clean runtime, erase "min"
```

```python
import regex as re
def clean_runtime(x):
    
    x=re.sub('min', '', x)
    
    return x
    
```

```python
comp['Runtime']=comp['Runtime'].apply(clean_runtime)
```

```python
comp["Runtime"].unique()
```

Dealing with genre

```python
# Films have multiple genres to them, in order to reduce the groups, I will take only the first genre.
# I am applying same methodology for all categorical columns, just picking the first element, understood as everything
#that goes up until a comma.
```

```python
comp["Genre"].unique()
```

```python
# I am doing this in two steps:
# Firstly, for the films that have three genres, I take out last two of them.
```

```python
import regex as re
def clean_genre(x):
    
    x=re.sub(',\W+\w*', '', x)
    
    return x
```

```python
comp['Genre']=comp['Genre'].apply(clean_genre)
```

```python
comp['Genre'].unique()
```

```python
#language
comp['language'].unique()
```

```python
def clean_language(x):
    
    x=re.sub(',\W+\w*', '', x)
    
    return x
comp['language']=comp['language'].apply(clean_language)
comp["language"].unique()
```

```python
#country
```

```python
comp['country'].unique()
```

```python
def clean_country(x):
    
    x=re.sub(',\W+\w*', '', x)
    
    return x
comp['country']=comp['country'].apply(clean_country)
comp["country"].unique()
```
